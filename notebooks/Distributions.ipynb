# Distributions

```python
import os
import glob
```

```python
# Previous to filtering
samples = glob.glob("03-Blast/*")
dists = {}
count = 0
for i in samples:
    print(f"Processing: {i}")
    files = glob.glob(f"{i}/*")
    d = {}
    for j in range(0,len(files)):
        # Progress
        if j % 200 == 0:
            print(f"{j}/{len(files)}")
        # Read file
        f = open(files[j],"r")
        for line in f:
            l = line.split("\t")[3]
            # Append new or sum to previous
            if l not in d.keys():
                d[l] = 1
            else:
                d[l] = d[l]+1
        f.close()
    # Append
    dists[i] = d
```
